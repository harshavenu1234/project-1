```python
from IPython.display import display, Audio
import subprocess
import os
import time
# List of file paths for your songs
file_paths = ["/content/sample_data/file example 1.mp3", "/content/file example2.mp3", "/content/file example 3.mp3"]
# Iterate over each file path
for file_path in file_paths:
    # Display and play the audio
    display(Audio(file_path, autoplay=True))
    # Get the duration of the audio file using ffprobe
    duration = float(subprocess.check_output(['ffprobe', '-i', file_path, '-show_entries', 'format=duration', '-v', 'quiet', '-of', 'csv=%s' % ("p=0")]))
    # Wait for the duration of the audio file before playing the next one
    time.sleep(duration)
```
